```python
import pandas as pd
from pandas import Series, DataFrame
import sqlite3
%matplotlib inline
from astropy.table import Table, Column
import numpy as np
```

```python
data = pd.read_csv("spend.csv")
data.head()
```

```python
inter_charge = data[data["Character Code"] == "57 - INTER DEPART CHARGES"]
inter_bill = data[data["Character Code"] == "58 - INTER DEPART BILLING"] 
```

```python
#which departments send the most bills to other departments 
bill_ag = inter_bill["Agency"].value_counts()
bill_ag.plot.bar(title="Frequency of Inter-Agency Billing", color="green")
```

```python
#which departments are charged the most by other departments
charge_ag = inter_charge["Agency"].value_counts()
charge_ag.plot.bar(title="Frequency of Inter-Agency Charges", color="green")
```
