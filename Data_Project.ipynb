```python
# filtering by two Organization

conn = sqlite3.connect('df.db')
df.to_sql('result', conn, if_exists = 'replace', index = False)
df_app = pd.read_sql('SELECT * from spend where organization =="17100 - IT APPLICATION DEVELOP SUPPORT" and agency == "17 -  INFORMATION TECHNOLOGY"', conn)
df_service = pd.read_sql('SELECT * from spend where organization =="17200 - IT TECHNICAL SERVICES" and agency == "17 -  INFORMATION TECHNOLOGY"', conn)
df_service
```

```python
# Two different Organization money spending
df_app = pd.DataFrame(df_app)
totals_app = []
years = ['2015', '2016', '2017', '2018']
for i in years:
    total= df_app[i].astype(float).sum()
    totals_app.append(total)
    
df_service = pd.DataFrame(df_service)
totals_service = []
years = ['2015', '2016', '2017', '2018']
for i in years:
    total= df_service[i].astype(float).sum()
    totals_service.append(total)  
    
df_totals_app #total spending in application
```

```python
# Two different Organization'spending
df_totals_service #total spending in service
```

```python
# Two different Organization'spending
df_totals_app = pd.Series(totals_app, years)
df_totals_service = pd.Series(totals_service, years)
df_concat = pd.concat([df_totals_app, df_totals_service], axis=1)

df_concat.T.plot.bar() 
ax = plt.gca()
ax.set_xticklabels(["IT APPLICATION DEVELOP SUPPORT", "IT TECHNICAL SERVICES"])
plt.title("Two organizations' spending per year");
```
